```python
book = book.rename({'User.ID': 'userid', 'Book.Title': 'title', 'Book.Rating': 'rating'}, axis = 1)
book.head()
book_d = book.drop(['Unnamed: 0'], axis = 1)
book_d = book_d.sort_values(['userid'])
```

```python
#number of unique users in the dataset
len(book_d.userid.unique())
```

```python
#number of unique titles in the dataset
len(book_d.title.unique())
```

```python
#number of unique ratings in the dataset
len(book_d.rating.unique())
```

```python
user_book= book_d.pivot_table(index='userid',columns='title',values='rating').reset_index()
```

```python
user_book
```

```python
user_book.index = book.userid.unique()
user_book
```

```python
#Impute those NaNs with 0 values
user_book.fillna(0, inplace=True)
user_book.head()
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_book.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
rbook = pd.DataFrame(user_sim)
```

```python
rbook.index = book_d.userid.unique()
rbook.columns = book_d.userid.unique()
```

```python
rbook.iloc[0:10,0:10]
```
